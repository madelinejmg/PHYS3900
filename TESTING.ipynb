### Reading KORG's ```galah_dr3_linelist.h5```

````python
import pandas as pd
import h5py
````

````python
ls -lh galah_dr3_linelist.h5
````

````python
dictionary = {}
with h5py.File('galah_dr3_linelist.h5', 'r') as f:
    for key in f.keys():
        print(f"Dataset: {key}, Shape: {f[key].shape}, Type: {f[key].dtype}")
        
        ds_arr = f[key][()]
        dictionary[key] = list(ds_arr)

df = pd.DataFrame(dictionary)
df
````

````python
with pd.HDFStore('galah_dr3_linelist.h5') as store:
    print(store.keys())
````
